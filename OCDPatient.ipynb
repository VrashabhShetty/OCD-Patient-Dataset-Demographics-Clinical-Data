**HANDLING OF MISSING DATA**

```python
# Find the mode of 'Previous Diagnoses' column
previous_diagnoses_mode = df['Previous Diagnoses'].mode()[0]

# Find the mode of 'Medications' column
medications_mode = df['Medications'].mode()[0]

# Fill missing values with the mode (without inplace=True)
df['Previous Diagnoses'] = df['Previous Diagnoses'].fillna(previous_diagnoses_mode)
df['Medications'] = df['Medications'].fillna(medications_mode)

# Display the updated DataFrame
df
```

**TO REMOVE LEADING WHITESPACES**

```python
##Remove leading whitespace in columns: 'Education Level', 'OCD Diagnosis Date' and 7 other columns
df['Education Level'] = df['Education Level'].str.lstrip()
df['OCD Diagnosis Date'] = df['OCD Diagnosis Date'].str.lstrip()
df['Previous Diagnoses'] = df['Previous Diagnoses'].str.lstrip()
df['Family History of OCD'] = df['Family History of OCD'].str.lstrip()
df['Obsession Type'] = df['Obsession Type'].str.lstrip()
df['Compulsion Type'] = df['Compulsion Type'].str.lstrip()
df['Depression Diagnosis'] = df['Depression Diagnosis'].str.lstrip()
df['Anxiety Diagnosis'] = df['Anxiety Diagnosis'].str.lstrip()
df['Medications'] = df['Medications'].str.lstrip()
```

**Gender column is converted into Numeric Values**

```python
#replacing the values in the 'Gender' column with 1 if the value is 'female' and 0 otherwise

df['Gender'] = df['Gender'].apply(lambda x: 1 if x.lower() == 'female' else 0)
df['Gender']
```

**Converting the categorical 'Previous diagnosis ', 'Education levels' , 'Ethnicity' and 'Marital status' into numerical values.**

```python
# Define the mapping
previous_diagnoses_mapping = {
    'MDD': 0,
    'GAD': 1,
    'Panic Disorder': 2,
    'PTSD': 3
}

# Apply the mapping to the 'Previous Diagnoses' column
df['Previous Diagnoses'] = df['Previous Diagnoses'].map(previous_diagnoses_mapping)
```

```python
# Define the mapping
education_level_mapping = {
    'High School': 0,
    'Some College': 1,
    'College Degree': 2,
    'Graduate Degree': 3
}

# Apply the mapping to the 'Education Level' column
df['Education Level'] = df['Education Level'].map(education_level_mapping)
```

```python
# Define the mapping
ethnicity_mapping = {
    'Caucasian': 0,
    'Asian': 1,
    'African': 2,
    'Hispanic': 3
}

# Apply the mapping to the 'Ethnicity' column
df['Ethnicity'] = df['Ethnicity'].map(ethnicity_mapping)
```

```python
# Define the mapping
marital_status_mapping = {
    'Single': 0,
    'Married': 1,
    'Divorced': 2
}

# Apply the mapping to the 'Marital Status' column
df['Marital Status'] = df['Marital Status'].map(marital_status_mapping)
```

**The distribution of Obsession Types across 2 genders**

```python
# Create the bar plot
plt.figure(figsize=(10, 5))
sns.set(style="whitegrid")

# Create the count plot with specified palette
plot = sns.countplot(data=df, x='Gender', hue='Obsession Type', palette='Oranges')

# Set the background color
sns.set(rc={'axes.facecolor':'lightblue', 'figure.facecolor':'lightblue'})
#Add a title by
plt.title('Obsession Type by Gender')
# Add data labels
for p in plot.patches:
    plot.annotate(format(p.get_height(), '.0f'),
                  (p.get_x() + p.get_width() / 2., p.get_height()),
                  ha = 'center', va = 'center',
                  xytext = (0, 10),
                  textcoords = 'offset points')

# Show the plot
plt.show()
```

**Obsession and Compulsion type with reference To Age**

```python
plt.figure(figsize=(10, 5))
sns.swarmplot(data=df, x='Obsession Type', y='Age', hue='Compulsion Type', palette='icefire')

# Add a title
plt.title('Obsession and Compulsion Types by Age')

# Add axis titles
plt.xlabel('Obsession Type')
plt.ylabel('Age')

# Add a legend
plt.legend(title='Compulsion Type', title_fontsize='13', loc='upper right')

# Display the plot
plt.show()
```

**Distribution of Patient's Education Level**

```python
# Count the occurrences of each education level
education_counts = df['Education Level'].value_counts().reset_index()
education_counts.columns = ['Education Level', 'Count']
# Create a DataFrame suitable for heatmap
heatmap_data = pd.DataFrame({
    'Education Level': education_counts['Education Level'],
    'Count': education_counts['Count']
})
#  Create the heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(heatmap_data.set_index('Education Level').T, annot=True, fmt="d", cmap='YlGnBu', cbar=False)
# Title and labels
plt.title("Heatmap of Patient's Education Level Distribution")
plt.xlabel('Education Level')

#  Display the heatmap
plt.show()
```

**Distribution of Patient's Ethnicity**

```python
# Assuming 'df_original' was the original DataFrame
# Create a copy of the original DataFrame before any modifications
# Create a copy of the original DataFrame, assuming it was named 'original_df'
# Replace 'original_df' with the actual name if it's different
original_df = pd.read_csv("OCD Patient Dataset_ Demographics & Clinical Data.csv")
df_original = original_df.copy()

# Calculate the counts of each ethnicity
ethnicity_counts = df_original['Ethnicity'].value_counts()

# Create a pie chart
plt.figure(figsize=(12, 6))
plt.pie(ethnicity_counts, labels=ethnicity_counts.index, autopct='%1.1f%%', startangle=90)

# Draw a circle in the center to make it a donut chart
centre_circle = plt.Circle((0, 0), 0.70, fc='skyblue')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)

# Equal aspect ratio ensures that the donut is circular
plt.axis('equal')

# Add a title
plt.title("Distribution of Patient's Ethnicity")

# Display the donut chart
plt.show()
```
